```python
# I want to add columns for metric of Test_success in the dataset and also the cumulative test scores.
# Since the maximum test scores for all the scores is 100.The median test score for each subject is 50.
# The cumulative passing metric will be 150
```

```python
df['cumulative_test_scores'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['Result'] = df['cumulative_test_scores'].apply(lambda x: 'Pass' if x >= 150 else 'Fail')
```

```python
df.head()
```

```python
df['Result'].value_counts()
```

```python
df = pd.get_dummies(df,drop_first=True)
```

```python
df.head()
```

**Modelling**

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['Result_Pass'],axis=1)
y = df['Result_Pass']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state =42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression()
```

```python
lr_model.fit(X_train,y_train)
```

```python
y_pred = lr_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
cnf_matrix = confusion_matrix(y_test,y_pred)
```
